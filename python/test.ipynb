```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
a = ListNode(1)
a.next = ListNode(2)
a.next.next = ListNode(3)
```

```python
temp = a
```

```python
def show_all_value(head):
    while head is not None:
        print(head.val)
        head = head.next
```

```python
temp = temp.next
```

```python
show_all_value(a)
```

```python
show_all_value(temp)
```
